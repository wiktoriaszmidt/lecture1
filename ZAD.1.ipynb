```python
import pandas as pd
import matplotlib.pyplot as plt

file_path = 'Salary Data (2).csv'
data = pd.read_csv(file_path)


avg_salary_by_education = data.groupby('Education Level')['Salary'].mean()


plt.figure(figsize=(10, 6))
avg_salary_by_education.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Średnie wynagrodzenie w zależności od poziomu wykształcenia')
plt.xlabel('Poziom wykształcenia')
plt.ylabel('Średnie wynagrodzenie')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
